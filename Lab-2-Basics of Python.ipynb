### **2. Variables in Python**  

A variable in Python is a named reference to a value stored in memory. It serves as a container for data, allowing values to be stored, modified, and used throughout a program.  

#### **Variable Nomenclature:**  
- Must begin with a letter or an underscore (`_`).  
- Can only contain letters, digits, and underscores.  
- Cannot be a reserved Python keyword (Check reserved keywords using `keyword.kwlist`).  
- Case-sensitive (e.g., `AuditData` and `auditdata` are different variables).  
- Avoid spaces in variable names; use underscores for readability (e.g., `total_revenue`).  

#### **Key Concepts to Remember:**  
- **Reference-based Storage**: Python variables do not store actual values but reference objects in memory.  
- **Dynamic Typing**: Python automatically assigns a data type at runtime, so explicit type declaration is not required.  
- **Memory Management**:  
  - When an **immutable object** (like `int`, `float`, `str`, `tuple`) is modified, a new object is created, and the variable reference is updated.  
  - When a **mutable object** (like `list`, `dict`, `set`) is modified, the reference remains the same, and the changes are made in place.  



Valid Variable Names (Starting with a Letter or Underscore)

```python
# Starts with a letter
audit_report = "Q1 Financial Audit"
transaction_count = 1200

# Starts with an underscore
_total_revenue = 5_000_000
_audit_risk_score = 92

# Printing values
print("Audit Report:", audit_report)
print("Transaction Count:", transaction_count)
print("Total Revenue:", _total_revenue)
print("Audit Risk Score:", _audit_risk_score)
```

 Using underscores at the beginning is useful for indicating internal or temporary variables.

Invalid Examples

```python
4total_expenditure = 150000  # ❌ Starts with a number
@audit_findings = 25         # ❌ Starts with a special character (@)
```

Variable names must always begin with a letter (A-Z, a-z) or an underscore (_). They cannot start with a digit or special characters.

**Addition (+)**

Calculating Total Revenue

Adding revenue from different branches to determine overall income.

```python
branch_1_revenue = 500000
print('Revenue from branch I:',branch_1_revenue)
branch_2_revenue = 350000
print('Revenue from branch II:',branch_2_revenue)
total_revenue = branch_1_revenue + branch_2_revenue
print("Total Revenue:", total_revenue)
```

**Subtraction (-)**

Determining Profit or Loss

Subtracting total expenses from revenue to compute net profit.

```python
total_revenue = 850000
print('Total Revenue:',total_revenue)
total_expenses = 600000
print('Total Expenses:',total_expenses)
net_profit = total_revenue - total_expenses
print("Net Profit:", net_profit)
```

**Multiplication (*)**

Calculating Interest on Audit Penalty

Applying an interest rate to an overdue amount to determine penalties.

```python
overdue_amount = 100000
print("Overdue amount(in INR):", overdue_amount)
interest_rate = 0.05  # 5% interest
print("Interest Rate (5% interest):", interest_rate)
total_penalty = overdue_amount * interest_rate
print("Total Penalty(in INR):", total_penalty)
```

**Division (/)**

Splitting Audit Costs Equally

Dividing total audit costs among departments for fair allocation.

```python
total_audit_cost = 120000
print("Total Audit Cost(in INR):", total_audit_cost)
departments = 4
print("Number of Departments", departments)
cost_per_department = total_audit_cost / departments
print("Cost per Department(in INR):", cost_per_department)
```

**Floor Division (//)**

Rounding Down Financial Estimates

Allocating assets equally across multiple branches, ensuring whole numbers.

```python
total_assets = 1050000
print("Total Value of Assets(in INR):", total_assets)
branches = 3
print("No. of Branches:", branches)
assets_per_branch = total_assets // branches
print("Assets Allocated Per Branch(in INR):", assets_per_branch)
```

**Exponentiation ()**

Calculating Compound Interest**

Applying interest growth over time using exponentiation.

```python
principal_amount = 50000
print("Principal Amount(in INR):", round(principal_amount, 2))
rate_of_interest = 0.08  # 8% interest
print("Interest Rate (8% interest):", rate_of_interest)
years = 5
final_amount = principal_amount * (1 + rate_of_interest) ** years
print("Final Amount After Interest(in INR):", round(final_amount, 2))
```

### **4. Logical Operators:**
Logical operators in Python (and, or, not) are used to combine multiple conditions and control the flow of decision-making. They help in writing efficient, readable, and structured code for complex conditions.

*   and Operator
*   or Operator
*   not Operator

**and** Operator

Ensuring Auditor Eligibility

The and operator returns True only if both conditions are met.

```python
# Check if an auditor has both a valid license and more than 5 years of experience.
auditor_license = True
print("Does the Auditor have a valid license?: ", auditor_license)
experience_years = 6
print("Years of experience as an Auditor: ", experience_years)
# Using 'and' to determine eligibility
auditor_eligibility = auditor_license and experience_years > 5
print("Auditor is eligible:", auditor_eligibility)
```

Ensures that an auditor is both certified and experienced before assigning critical tasks.

**or** Operator

Flagging Suspicious Transactions

The or operator returns True if at least one condition is met.

```python
# A transaction is flagged if it is above $10,000 or if it comes from an unverified account.
transaction_amount = 8000
print('Transaction amount(in $): ', transaction_amount)
is_verified_account = False
print('Is the account verified?: ', is_verified_account)

# Using 'or' to check if it should be flagged
flagged = transaction_amount > 10000 or not is_verified_account
print("Transaction flagged:", flagged)
```

Helps flag high-risk transactions for further review.

**not** Operator

Identifying Missing Financial Data

The not operator inverts a boolean value (True becomes False, and vice versa).

```python
# If the financial data is not available, print a warning.
financial_data_available = False
print("Is the financial data available:", financial_data_available)

# Using 'not' to check data status
print("Financial data missing:", not financial_data_available)
```

Alerts auditors if financial records are incomplete.

### **5. Membership Operators in Python**

Membership operators in and not in are used to check whether a value exists in a sequence (like a list, tuple, dictionary, or string).


*   (in) operator
*   (not in ) operator

**in** Operator

Checking Audited Transactions

The in operator checks if a value exists within a sequence.

```python
# List of audited transaction IDs
audited_transactions = ["TXN1001", "TXN1002", "TXN1003", "TXN1004"]
print("Audited Transaction IDs:")
for tran_ID in audited_transactions:
    print(tran_ID)


# Checking if a specific transaction was audited
transaction_to_check = "TXN1002"
print("\nHas Transaction ID:",transaction_to_check,"been audited? :", transaction_to_check in audited_transactions)
```

Helps auditors verify whether a transaction was already reviewed.

**not in** Operator

Verifying Blacklisted Companies

The not in operator checks if a value does not exist in a sequence.

```python
# Blacklisted companies
blacklisted_companies = ["ABC Corp", "XYZ Ltd", "Fraud Inc"]
print("Blacklisted companies:")
for bl_company in blacklisted_companies:
    print(bl_company)

# Checking if a company is NOT blacklisted
company_name = "Safe Enterprises"
print("\nCompany '",company_name,"' is NOT in the blacklisted company list:", company_name not in blacklisted_companies)
```

Ensures that a company is not on a fraud watchlist before processing transactions.

### **6.Identity Operators:**
Identity operators in Python (is, is not) are used to compare memory locations of objects, rather than just values. They are useful in audit systems for validating data consistency, checking object identity, and ensuring records are correctly referenced.

*   (is) operators
*   (is not) operators

is Operator – Checking If Two Audit Records Refer to the Same Object
The is operator checks if two variables refer to the same memory location.

```python
# Two audit reports assigned to the same object
audit_report_1 = {"year": 2024, "status": "Pending"}
print('Audit Report I:',audit_report_1)
audit_report_2 = audit_report_1  # Both point to the same object
print('Audit Report II:',audit_report_2)

# Checking identity
print("Both refer to the same audit report?:", audit_report_1 is audit_report_2)
```

Ensures that two variables are actually referencing the same object and not just identical data.

is not Operator – Checking If Two Audit Records Are Different
The is not operator checks if two variables do not refer to the same object in memory.




```python
# Two separate audit reports
audit_report_1 = {"year": 2024, "status": "Pending"}
print('Audit Report I:',audit_report_1)
audit_report_2 = {"year": 2024, "status": "Pending"}  # Different object
print('Audit Report II:',audit_report_2)
# Checking identity
print("Audit reports are different:", audit_report_1 is not audit_report_2)
```

Useful in auditing systems where you need to confirm that two reports are distinct objects, even if they contain identical data.

### **7.Bitwise Operators:**



*   AND (&)

*   OR (|)
*   XOR(^)


*   NOT (~)




**Bitwise AND (&)**

Checking User Permissions

The Bitwise AND (&) operator checks if both bits are 1.

```python
# Permissions stored as binary flags (1 = granted, 0 = denied)
READ = 0b01   # Read access with Binary: 01
WRITE = 0b10  # Write access with Binary: 10

user_permission = 0b11  # User has both read (01) and write (10) permissions

# Checking if the user has both permissions using Bitwise AND
has_read_write = user_permission & (READ | WRITE)

print("User has both read and write access:", has_read_write == (READ | WRITE))
```

Ensures a user has the required access to a financial system.

**Bitwise OR (|)**

Granting Additional Permissions

The Bitwise OR (|) operator adds permissions while keeping existing ones.

```python
# An auditor updates a user's access level by granting write permissions.

#User's initial permissions (Binary: 01) - Read-only access
user_permission = 0b01  # Read permission (bit 0 set)
print("Initial Permission (Binary):", bin(user_permission))

# Grant write permission using bitwise OR (Binary: 10)
user_permission |= 0b10  # Now user has both read (01) and write (10) permissions
print("Updated Permission (Binary):", bin(user_permission)) # Read + Write
```

Used in role-based access control (RBAC) to add permissions.

**Bitwise XOR (^)**

Detecting Financial Record Alterations

The Bitwise XOR (^) operator detects differences between two binary values.



```python
# An auditor detects if financial records have been altered by comparing two copies of a file.

original_record = 0b1101  # Original financial data (binary)
print("Original Record(in decimal):",original_record)
tampered_record = 0b1001  # Possibly altered data
print("Record allegedly tampered with(in decimal):",tampered_record)

# XOR will return nonzero if there's a difference
difference = original_record ^ tampered_record
print("Difference in record value:", difference)

print("Data tampered:", difference != 0)  # Output: True (Data has been altered)
```

Useful in data integrity checks by comparing two financial records.

**Bitwise NOT (~)**

Toggling Transaction Status

The Bitwise NOT (~) operator flips all bits, helping in flag manipulations.


```python
# A bitwise NOT operation can be used to toggle flags, such as marking a transaction as fraudulent.
transaction_status = 0b0001  # Transaction is marked as "Valid"
print('Transaction Status(in decimal):', transaction_status)

fraud_status = ~transaction_status  # Reverse the flag (Toggle status)
#### ~x = -(x + 1) Therefore => ~1 = -(1 + 1) = -2
print('Initial fraud status(Bitwise NOT of 1):', fraud_status)

print("Fraud Status (Binary):", bin(fraud_status & 0b1111))  # Output: 0b1110 (Marked as Fraud)
```

Helps in flag-based security checks like fraud detection.

### **Order of Operations**

When conducting financial audits, tax evaluations, or expense verifications, auditors must apply Python's order of operations (PEMDAS) correctly to ensure accurate calculations.

**PEMDAS Rule:**

P - Parentheses () → Solve expressions inside parentheses first.

E - Exponents ** → Apply power calculations.

MD - Multiplication * & Division /, //, % → Evaluated left to right.

AS - Addition + & Subtraction - → Evaluated left to right.

**Example 1 : Expense Calculation for an Audit**

An auditor is analyzing a company’s total expense after considering:

- Discount on initial expense

- Taxation impact over multiple years

- Depreciation effects

```python
# Given financial data
initial_expense = 1000  # Expense before tax & discounts ($)
tax_rate = 0.08         # 8% tax applied annually
discount = 0.1          # 10% discount
years = 5               # Depreciation over 5 years
depreciation_rate = 0.05  # 5% yearly depreciation

print('Initial Expense before tax & discounts ($):', initial_expense)
print('Tax Rate Applied(in %):', tax_rate*100)
print('Discount Rate Applied(in %):', discount*100)
print('Depreciation Rate Applied(in %):', depreciation_rate*100)


# Applying PEMDAS for correct calculation
final_expense = (initial_expense * (1 - discount)) * ((1 + tax_rate) ** years) - (initial_expense * depreciation_rate * years)

# Display the final computed expense
print(f"The final expense after tax, discount, and depreciation over {years} years is: ${final_expense:.2f}")
```
